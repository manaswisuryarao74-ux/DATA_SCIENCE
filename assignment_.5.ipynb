# side-by-side Bar Chart

```python
import matplotlib.pyplot as mp
import numpy as np

years = ['2001', '2006', '2010', '2015', '2018']
sachin_runs = [1600, 1300, 1562, 0, 0]
kohli_runs = [0, 0, 995, 1380, 1325]
sehwag_runs = [850, 900, 1100, 0, 0]

x = np.arange(len(years))
width = 0.25

mp.figure(figsize = (10,6))
mp.bar(x - width, sachin_runs, width, label = 'Sachin')
mp.bar(x,kohli_runs, width, label = 'Kohli')
mp.bar(x + width, sehwag_runs, width, label = 'Sehwag')

mp.xlabel('YEARS')
mp.ylabel('RUNS')
mp.title('Over 5 selected Years')
mp.xticks(x, years)

mp.legend()
mp.show()
```

```python
players = ['sachin', 'kohli', 'sehwag']
debut_runs = [4820,4886,4295]

mp.figure(figsize = (8,4))
mp.barh(players, debut_runs, color=['orange','pink','red'])
mp.xlabel('Total Runs')
mp.title('Total Runs in Debut 5 Years')

mp.tight_layout()
mp.show()
```

# PIE Chart

```python
labels = ['Android', 'iOS', 'Others']
market_share = [72, 10, 2]
explode = [0.2, 0, 0] 

mp.figure(figsize=(6,6))
mp.pie(market_share, labels=labels, autopct='%1.1f%%', startangle=140, shadow=True, explode=explode)
mp.title("Mobile OS Market Share - Pie Chart")
mp.show()
```

```python
labels = ['Android', 'iOS', 'Others']
market_shares = [72, 10, 2]
mp.barh(labels, market_shares, color=['red','orange','gray'])
mp.xlabel("Market Share (%)")
mp.title("Mobile OS Market Share (Horizontal Bar)")

mp.tight_layout()
mp.show()
```
