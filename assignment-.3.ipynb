# Numpy Function

```python
import numpy as man
arr = man.array([1,2,3,4])
print(arr)
```

```python
# 1. SORT
```

```python
arr1 = man.array([43,5,23,7,10,1]) #sorting array
print("sorted array :",man.sort(arr1))
```

```python
# 2. CLIP
```

```python
man.clip(arr,1,2) # limits the values in an array to a specified range.

```

```python
# 3. FLOOR
```

```python
flo = man.array([2.3,3.4,5.8,34.6,5.1,23.0]) #returns the largest integer less than or equal to each element in an array.
print("Floor of array :",man.floor(flo))
```

```python
# 4. ROUND
```

```python
rou = man.array([22.433,43.5654,67.7867,34.243]) #nearest integer
print("Round of array :",man.round(rou))
```

```python
rou = man.array([22.433,43.5654,67.7867,34.243])
print("Round of array :",man.round(rou,1))
```

```python
rou = man.array([22.433,43.5654,67.7867,34.243])
print("Round of array :",man.round(rou,2))
```

```python
# 5.CEIL
```

```python
print("CEIL :",man.ceil(flo)) # smallest integer in an array
```

```python
# 6.INTERSECT
```

```python
a = man.array([1,2,3]) 
b = man.array([2,4,5])
print("intersect array :",man.intersect1d(a,b)) #return unique value that are present in both array
```

```python
# 7.UNION
```

```python
print("union array :",man.union1d(a,b)) #Unique values means repated value print nahi kart
```

```python
# 8.SET DIFFERENCE
```

```python
print("set difference array :",man.setdiff1d(a,b))
```

```python
print("set difference array :",man.setdiff1d(b,a))
```
